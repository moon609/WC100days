### 2. å–å‡ºæ¯ä¸€ç­†è³‡æ–™çš„ã€Œæ¨™é¡Œã€ã€ã€Œè²¼æ–‡æ™‚é–“ã€ã€ã€Œç•™è¨€äººæ•¸ã€ã€ã€ŒæŒ‰è®šäººæ•¸ã€

```python
# ç”¨è¿´åœˆå­˜å–æ¯ä¸€ç­†è³‡æ–™çš„ã€Œæ¨™é¡Œã€ã€ã€Œè²¼æ–‡æ™‚é–“ã€ã€ã€Œç•™è¨€äººæ•¸ã€ã€ã€ŒæŒ‰è®šäººæ•¸ã€
for i in range(len(j)): 
    print (j[i]['title'], #æ¨™é¡Œ
           j[i]['createdAt'], #è²¼æ–‡æ™‚é–“
           j[i]['commentCount'], #ç•™è¨€äººæ•¸
           j[i]['likeCount']) #æŒ‰è®šäººæ•¸
```

### 3. è¨ˆç®—ç†±é–€/éç†±é–€æ–‡ç« çš„ã€Œå¹³å‡ç•™è¨€äººæ•¸ã€èˆ‡ã€Œå¹³å‡æŒ‰è®šäººæ•¸ã€

```python
# è®€å–ç†±é–€æ–‡ç« çš„è³‡æ–™
r = requests.get('https://www.dcard.tw/_api/forums/pet/posts?popular=true') #è¨­å®špopularç‚ºtrue
j=json.loads(r.text)

#æŠ“å–ç†±é–€æ–‡ç« çš„ã€Œç•™è¨€äººæ•¸ã€èˆ‡ã€ŒæŒ‰è®šäººæ•¸ã€
hot_comm=[]
hot_like=[]
for i in range(len(j)): 
    hot_comm.append(j[i].get('commentCount'))
    hot_like.append(j[i].get('likeCount'))
print (hot_comm)
print (hot_like)

#è¨ˆç®—å¹³å‡
print("ç†±é–€æ–‡ç« çš„ã€Œå¹³å‡ç•™è¨€äººæ•¸ã€ç‚º: " + str(sum(hot_comm)/len(hot_comm)) + "äºº") 
print("ç†±é–€æ–‡ç« çš„ã€Œå¹³å‡æŒ‰è®šäººæ•¸ã€ç‚º: " + str(sum(hot_like)/len(hot_like)) + "äºº") 
```

```python
# è®€å–éç†±é–€æ–‡ç« çš„è³‡æ–™
nonr = requests.get('https://www.dcard.tw/_api/forums/pet/posts?popular=false') #è¨­å®špopularç‚ºfalse
nonj=json.loads(nonr.text)

#æŠ“å–éç†±é–€æ–‡ç« çš„ã€Œç•™è¨€äººæ•¸ã€èˆ‡ã€ŒæŒ‰è®šäººæ•¸ã€
nonhot_comm=[]
nonhot_like=[]
for i in range(len(nonj)): 
    nonhot_comm.append(nonj[i].get('commentCount'))
    nonhot_like.append(nonj[i].get('likeCount'))
print (nonhot_comm)
print (nonhot_like)

#è¨ˆç®—å¹³å‡
print("éç†±é–€æ–‡ç« çš„ã€Œå¹³å‡ç•™è¨€äººæ•¸ã€ç‚º: " + str(sum(nonhot_comm)/len(nonhot_comm)) + "äºº") 
print("éç†±é–€æ–‡ç« çš„ã€Œå¹³å‡æŒ‰è®šäººæ•¸ã€ç‚º: " + str(sum(nonhot_like)/len(nonhot_like)) + "äºº") 
```
